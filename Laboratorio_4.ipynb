### **b. Detección y eliminación de outliers univariados en una columna usando el método del rango intercuartílico (IQR) con un umbral de 3 veces el IQR a la izquierda y a la derecha.**

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import mahalanobis
from numpy.linalg import inv
```

```python
# Eliminar la última columna vacía si existe
df.dropna(axis=1, how='all', inplace=True)
```

```python
# Detección de outliers univariados (Rango Intercuartílico con factor 3)
numeric_df = df.drop(columns=[0, 1])  # Excluir ID y diagnóstico

Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 3 * IQR
upper_bound = Q3 + 3 * IQR

mask = ~((numeric_df < lower_bound) | (numeric_df > upper_bound)).any(axis=1)
df_iqr_cleaned = df[mask].copy()
```

```python
# Detección de outliers multivariados (Distancia de Mahalanobis > 30)
numeric_data = df_iqr_cleaned.drop(columns=[0, 1]).values
mean_vec = np.mean(numeric_data, axis=0)
cov_matrix = np.cov(numeric_data, rowvar=False)
inv_cov_matrix = inv(cov_matrix)

mahalanobis_distances = [mahalanobis(row, mean_vec, inv_cov_matrix) for row in numeric_data]
df_iqr_cleaned["mahalanobis"] = mahalanobis_distances

df_cleaned = df_iqr_cleaned[df_iqr_cleaned["mahalanobis"] <= 30].drop(columns=["mahalanobis"])
```

```python
# Verificación final
print("Filas originales:", df.shape[0])
print("Después de eliminar outliers univariados:", df_iqr_cleaned.shape[0])
print("Después de eliminar outliers multivariados:", df_cleaned.shape[0])
```

### **c. Calcular estadísticos descriptivos para las variables predictoras.**

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
numeric_df = df.drop(columns=[0, 1])

Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

mask = ~((numeric_df < lower_bound) | (numeric_df > upper_bound)).any(axis=1)
df_iqr_cleaned = df[mask].copy()
```

```python
# Mahalanobis (igual que antes)
from scipy.spatial.distance import mahalanobis
from numpy.linalg import inv
import numpy as np
numeric_data = df_iqr_cleaned.drop(columns=[0, 1]).values
mean_vec = np.mean(numeric_data, axis=0)
cov_matrix = np.cov(numeric_data, rowvar=False)
inv_cov_matrix = inv(cov_matrix)

mahalanobis_distances = [mahalanobis(row, mean_vec, inv_cov_matrix) for row in numeric_data]
df_iqr_cleaned["mahalanobis"] = mahalanobis_distances
df_cleaned = df_iqr_cleaned[df_iqr_cleaned["mahalanobis"] <= 30].drop(columns=["mahalanobis"])
```
