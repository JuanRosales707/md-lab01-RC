### **b. Realice una detección de valores atípicos univariados por medio del método del rango intercuartílico con 3 de longitud a la derecha y 3 a la izquierda, y elimínelos. Además, realiceuna detección de valores atípicos multivariados por medio de las distancias de Mahalanobis y elimine aquellos valores que superen el valor de 30.**

```python
from scipy.spatial.distance import mahalanobis
from scipy.stats import zscore
import numpy as np
```

```python
# Copiamos el DataFrame original (asumiendo que 'df' ya fue limpiado antes)
df_iqr = df.copy()
# Quitamos la columna 0 ('ID') para no analizarla
df_iqr_numeric = df_iqr.drop(columns=[0])
```

```python
# Convertir todas las columnas excepto la columna 0 (ID) a tipo numérico
df_iqr_numeric = df_iqr.drop(columns=[0]).apply(pd.to_numeric, errors='coerce')
```

```python
# Calcular cuartiles e IQR
Q1 = df_iqr_numeric.quantile(0.25)
Q3 = df_iqr_numeric.quantile(0.75)
IQR = Q3 - Q1

# Crear condición para filtrar outliers
condition = ~((df_iqr_numeric < (Q1 - 3 * IQR)) | (df_iqr_numeric > (Q3 + 3 * IQR))).any(axis=1)

# Aplicar filtro al DataFrame original
df_clean_iqr = df_iqr[condition]

print(f"Filas después de limpieza univariada: {df_clean_iqr.shape[0]}")
```

```python
# Seleccionamos solo las columnas numéricas (sin ID)
df_mv = df_clean_iqr.drop(columns=[0]).apply(pd.to_numeric, errors='coerce')
# Calculamos el vector de medias y la matriz de covarianza inversa
mean_vec = df_mv.mean().values
cov_matrix = np.cov(df_mv.T)
inv_cov_matrix = np.linalg.inv(cov_matrix)
# Calculamos la distancia de Mahalanobis para cada fila
mahal_distances = df_mv.apply(lambda row: mahalanobis(row, mean_vec, inv_cov_matrix), axis=1)

# Filtramos filas donde la distancia es menor o igual a 30
df_final = df_clean_iqr[mahal_distances <= 30]
```

```python
# Mostramos resultado final
print(f"Filas después de limpieza multivariada (Mahalanobis): {df_final.shape[0]}")
```
